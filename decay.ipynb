```python
import pandas
import numpy as np
m=5000
N=400
R=0.0000000001
```

```python
def K(x,y):
    return np.dot(x,y)+(np.dot(x,y))**2
```

```python
trainx=pandas.read_csv('trainx1.csv',header=None)
```

```python
X_train=pandas.DataFrame(trainx).to_numpy()
trainy=pandas.read_csv('trainy1.csv',header=None)
Y_train=pandas.DataFrame(trainy).to_numpy()
print(X_train.shape)
```

```python
testx=pandas.read_csv('testx1.csv',header=None)
Xtest=pandas.DataFrame(testx).to_numpy()
```

```python
b=np.zeros((m,N))
for i in range(m):
    for j in range(N):
        b[i][j]=K(Xtest[i],X_train[j])
b=np.matrix(b) 
I=np.identity(N)
k=np.zeros((N,N))
for i in range(N):
    for j in range(N):
        k[i][j]=K(X_train[i],X_train[j])

k=np.matrix(k)+R*I
k_inv=np.linalg.inv(k)
y=Y_train
temp1=np.matmul(b,k_inv)
temp2=np.matmul(temp1,y)



```

```python
num=np.linalg.norm(temp2)
```
