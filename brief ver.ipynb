```python
import numpy as np
import pandas as pd
import matplotlib as plt
import EigenPro2.mnist as mnist
from sklearn import metrics
import math
import time
```

```python
(train_x, train_y), (test_x, test_y) = mnist.load()

```

```python
X_temp=[]
X=[]
Y=[]
Y_temp=[]
for i in range(60000):
    if train_y[i] ==2:
        
        temp=train_x[i].flatten()
        temp2=-1
        X_temp.append(temp)
        Y_temp.append(temp2)
    elif train_y[i]==5:
        temp=train_x[i].flatten()
        temp2=1
        X_temp.append(temp)
        Y_temp.append(temp2)


for i in range(5000):
    Y.append(Y_temp[i])
    X.append(X_temp[i])
```

```python
X_test=[]
Y_test=[]
for i in range(10000):
    if test_y[i] ==2:
        
        temp=train_x[i].flatten()
        temp2=-1
        X_test.append(temp)
        Y_test.append(temp2)
    elif test_y[i]==5:
        temp=train_x[i].flatten()
        temp2=1
        X_test.append(temp)
        Y_test.append(temp2)
y_bar=np.average(Y_test)
print(len(Y_test))
```
